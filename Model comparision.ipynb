```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = [
    "Mobile_wo_resize",
    "Mobile_optimized",
    "Efficient_frozen",
    "Efficient_optimized",
    "ResNet50_frozen",
    "ResNet50_optimized"
]

accuracies = [
    0.3571,
    0.8712,
    0.9114,
    0.9387,
    0.9021,
    0.9446
]

# Grouped data
groups = ["MobileNet", "EfficientNet", "ResNet"]
before = [accuracies[0], accuracies[2], accuracies[4]]
after = [accuracies[1], accuracies[3], accuracies[5]]

x = np.arange(len(groups))
width = 0.35

plt.figure(figsize=(10, 6))
bars_before = plt.bar(x - width/2, before, width, label="Before", color="lightcyan")
bars_after = plt.bar(x + width/2, after, width, label="After", color="skyblue")

# Annotate bars
for bar in bars_before:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f"{height:.3f}",
        ha="center",
        va="bottom"
    )

for bar in bars_after:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f"{height:.3f}",
        ha="center",
        va="bottom"
    )

plt.xticks(x, groups)
plt.ylabel("Accuracy")
plt.title("Model Accuracy Before and After Optimization")
plt.legend()
plt.tight_layout()

plt.show()
```

#### Model Comparision

 CNN baseline model : Test Accuracy: 0.6997
 CNN optimzied model : Test Accuracy: 0.8644
 MobileNetV2 : Test accuracy: 0.8713
 EfficientNetB0: Test accuracy: 0.9387
 ResNet50: Test Accuracy: 0.9446
 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Model names and accuracies
models = [
    "CNN Baseline",
    "CNN Optimized",
    "MobileNetV2",
    "EfficientNetB0",
    "ResNet50"
]

accuracies = [
    0.6997,
    0.8644,
    0.8713,
    0.9387,
    0.9446
]


df = pd.DataFrame({
    "Model": models,
    "Accuracy": accuracies
})


plt.figure(figsize=(10, 6))

ax = sns.barplot(
    data=df,
    x="Model",
    y="Accuracy",
    palette='flare',
    width=0.7
)

# Add percentage annotations (rounded to 2 decimals)
for p in ax.patches:
    height = p.get_height()
    pct = height * 100
    x = p.get_x() + p.get_width() / 2
    ax.annotate(
        f"{pct:.2f}%",
        (x, height),
        ha='center',
        va='bottom',
        fontsize=11,
        xytext=(0, 5),
        textcoords="offset points"
    )


sns.barplot(x=models, y=accuracies, data=df, palette='flare', legend=True)
plt.title("Test Accuracy Comparison", fontsize=16)
plt.ylabel("Accuracy")
plt.ylim(0, 1.0)
plt.xticks(rotation=30)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data
models = [
    "CNN Baseline",
    "CNN Optimized",
    "MobileNetV2",
    "EfficientNetB0",
    "ResNet50"
]

accuracies = [
    0.6997,
    0.8644,
    0.8713,
    0.9387,
    0.9446
]

df = pd.DataFrame({
    "Model": models,
    "Accuracy": accuracies
})

palette = ["red", "orange", "yellow", "green", "blue"]

plt.figure(figsize=(10, 6))

ax = sns.barplot(
    data=df,
    x="Model",
    y="Accuracy",
    palette=palette,
    width=0.7
)

# Add percentage annotations (rounded to 2 decimals)
for p in ax.patches:
    height = p.get_height()
    pct = height * 100
    x = p.get_x() + p.get_width() / 2
    ax.annotate(
        f"{pct:.2f}%",
        (x, height),
        ha='center',
        va='bottom',
        fontsize=11,
        xytext=(0, 5),
        textcoords="offset points"
    )

plt.ylim(0, 1.05)
plt.xticks(rotation=30)
plt.ylabel("Accuracy")
plt.title("Model Test Accuracy Comparison (Percent Labels)")

plt.tight_layout()
plt.show()
```
